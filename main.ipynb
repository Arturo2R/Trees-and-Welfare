```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Creamos una malla para los bienes/trabajo/capital
x = np.linspace(0.01, 1, 50)
y = np.linspace(0.01, 1, 50)
X, Y = np.meshgrid(x, y)

# Utilidades (intercambio) - funciones de utilidad para A y B
U_A = X**0.5 * Y**0.5
U_B = (1 - X)**0.5 * (1 - Y)**0.5

# Isocuantas de produccin - para dos bienes (asumimos tecnologa Cobb-Douglas)
Q1 = X**0.4 * Y**0.6
Q2 = (1 - X)**0.6 * (1 - Y)**0.4

# Frontera de posibilidades de produccin (simplificada)
FPP = 1 - X**2

# Funcin de utilidad social simplificada
U_social = X**0.5 * FPP**0.5

# Creamos el plot 3D
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# Superficie de utilidad individual (intercambio)
ax.plot_surface(X, Y, U_A, alpha=0.5, cmap='Blues', label='U_A')
ax.plot_surface(X, Y, U_B, alpha=0.5, cmap='Greens', label='U_B')

# Superficie de produccin (isocuantas de bienes)
ax.plot_surface(X, Y, Q1, alpha=0.3, cmap='Oranges')
ax.plot_surface(X, Y, Q2, alpha=0.3, cmap='Reds')

# Superficie de eficiencia asignativa
ax.plot_surface(X, FPP, U_social, alpha=0.6, cmap='Purples')

# Etiquetas y ttulo
ax.set_title('3D: Eficiencia de Pareto en Intercambio, Produccin y Asignacin', fontsize=14)
ax.set_xlabel('Bien X / Trabajo')
ax.set_ylabel('Bien Y / Capital')
ax.set_zlabel('Nivel de Utilidad / Produccin / Bienestar Social')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuramos el estilo


fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# --------------------
# Panel 1: Intercambio (Caja de Edgeworth)
# --------------------
axs[0].set_title("Eficiencia en el Intercambio")
axs[0].set_xlabel("Bien X")
axs[0].set_ylabel("Bien Y")

# Curvas de indiferencia
x = np.linspace(0.1, 1, 100)
for a in [0.2, 0.4, 0.6]:
    axs[0].plot(x, a / x, label=f"U = {a}", linestyle='--', color='blue')

for b in [0.2, 0.4, 0.6]:
    axs[0].plot(x, b / (1 - x + 0.001), linestyle='--', color='green')

# Curva de contrato
contract_curve_x = np.linspace(0.05, 0.95, 100)
contract_curve_y = contract_curve_x  # simplificacin
axs[0].plot(contract_curve_x, contract_curve_y, color='black', label="Curva de Contrato")
axs[0].legend()
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)

# --------------------
# Panel 2: Produccin (Caja de insumos)
# --------------------
axs[1].set_title("Eficiencia en la Produccin")
axs[1].set_xlabel("Trabajo")
axs[1].set_ylabel("Capital")

# Isocuantas para Bien X y Bien Y
x = np.linspace(0.1, 1, 100)
for q in [0.3, 0.5, 0.7]:
    axs[1].plot(x, q / x, linestyle='--', color='red')

for q in [0.3, 0.5, 0.7]:
    axs[1].plot(x, 1 - q / x, linestyle='--', color='orange')

# Curva de contrato de produccin (simplificada)
prod_contract_x = np.linspace(0.05, 0.95, 100)
prod_contract_y = 1 - prod_contract_x
axs[1].plot(prod_contract_x, prod_contract_y, color='black', label="Curva de Contrato de Produccin")
axs[1].legend()
axs[1].set_xlim(0, 1)
axs[1].set_ylim(0, 1)

# --------------------
# Panel 3: Asignacin final (FPP y bienestar)
# --------------------
axs[2].set_title("Eficiencia Asignativa (FPP y Bienestar Social)")
axs[2].set_xlabel("Cantidad de Bien X")
axs[2].set_ylabel("Cantidad de Bien Y")

# Frontera de posibilidades de produccin (FPP)
x = np.linspace(0, 1, 100)
fpp = 1 - x**2
axs[2].plot(x, fpp, label="FPP", color='black')

# Curva de bienestar social (simplificada como U_social = xy)
for u in [0.2, 0.3, 0.4]:
    axs[2].plot(x, u / (x + 0.001), linestyle='--', label=f"U_s = {u}", color='purple')

axs[2].legend()
axs[2].set_xlim(0, 1)
axs[2].set_ylim(0, 1.2)

plt.tight_layout()
plt.suptitle("Tres Condiciones de Eficiencia de Pareto", fontsize=16, y=1.05)
plt.show()
```
